**6. Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text**<br>
HINT: You should include an `'IS_SPACE': True` pattern between the two words!

```python
# Import the Matcher library:

from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

```python
# Create a pattern and add it to matcher:
swimming_1 = [{'LOWER': 'swimming'}, {'IS_SPACE' : True}, {'LOWER': 'vigorously'}];
swimming_2 = [{'ORTH': 'Swimming'}, {'ORTH': 'Vigorously'}];
#swimming_3 = {'LOWER': 'swim', 'IS_SPACE': True, 'LOWER': 'vigor'}
#swimming_1 = {'LOWER': 'swimming'};
#swimming_2 = {'LOWER': 'vigorously'};
pattern = [{'LOWER': 'swimming'}, {'IS_SPACE': True, 'OP':'*'}, {'LOWER': 'vigorously'}]
matcher.add('Swimmer', None, swimming_1);

```

```python
#nlp('swimming')[0].lemma_
matcher.get('Swimmer')
```

```python
# Create a list of matches called "found_matches" and print the list:
found_matches = matcher(doc);
print (found_matches)
#print (doc)
```
